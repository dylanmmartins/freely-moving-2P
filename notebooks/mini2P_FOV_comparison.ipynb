```python
fig, axs = plt.subplots(4,1, figsize=(5,6), dpi=300)
for fig_i, ops_path in enumerate(ops_paths):
    ops = np.load(ops_path, allow_pickle=True)
    axs[fig_i].hist(ops.item()['xoff'], alpha=0.5, label='x', bins=np.linspace(-7,7,16))
    axs[fig_i].hist(ops.item()['yoff'], alpha=0.5, label='y', bins=np.linspace(-7,7,16))
    axs[fig_i].set_xlim([-7,7])
for fig_i in range(3):
    axs[fig_i].set_xticks([])
axs[3].set_xlabel('image shift (pxls)')
axs[0].legend()
fig.suptitle('registration shift')
fig.tight_layout()
```

```python
fig, axs = plt.subplots(2,2, figsize=(9,8), dpi=300)
axs = axs.flatten()
for fig_i, ops_path in enumerate(ops_paths):
    ops = np.load(ops_path, allow_pickle=True)
    axs[fig_i].imshow(ops.item()['meanImg'], cmap='gray')
    axs[fig_i].axis('off')
    title = str(os.path.normpath(ops_path).split('\\')[2])
    if title == 'suite2p':
        title = str(os.path.normpath(ops_path).split('\\')[2])
    axs[fig_i].set_title(title)
```
