```python
import numpy as np
import matplotlib.pyplot as plt
import fm2p
```

```python
pvals = np.load(r'K:\Mini2P\250306_DMM_DMM038_pillar\fm2\cell_pvalues.npy')
```

```python
lag_vals = [-3,-2,-1,0,1,2,3,4,5]
```

```python
len(lag_vals)
```

```python
pvals[:,2,:]
```

```python
fig, axs = plt.subplots(3,3, dpi=300, figsize=(10,6))
axs = axs.flatten()
for lag_ind, lag_val in enumerate(lag_vals):
    axs[lag_ind].hist(pvals[:,0,lag_ind], color='tab:blue', bins=100)
    axs[lag_ind].set_xlabel('p value')
    axs[lag_ind].set_ylabel('cells')
    axs[lag_ind].set_title('lag={}f'.format(lag_val))
fig.suptitle('pupil')
fig.tight_layout()
```
